```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import string
import re
```

```python
#import dataset
df_tweet = pd.read_csv("TweetLabeling.csv", encoding = 'utf-8')
df_tweet
```

```python
#Menghilangkan kolom yang tidak dibutuhkan
df_tweet = df_tweet.loc[:, ~df_tweet.columns.str.contains('^Unnamed')]
df_tweet.drop('User', axis=1, inplace=True)
df_tweet.drop('Translate', axis=1, inplace=True)
df_tweet.drop('Sentiment', axis=1, inplace=True)
df_tweet
```

```python
#Mendefinisikan nilai hasil
#Untuk Positif = 1, Netral = 0 dan Negatif = -1
df_tweet['Hasil'] = df_tweet['Hasil'].map({"Positif" : 1, "Netral" : 0, "Negatif" : -1})
df_tweet
```

```python
df_tweet.groupby('Hasil').describe()
```

```python
#Proses Casefolding
def token(preprocessing):
  nstr = preprocessing.split(' ')
  dat = []
  a = -1
  for huruf in nstr:
    a = a + 1
  if huruf == '':
    dat.append(a)
  p = 0
  b = 0
  for q in dat:
    b = q - p
    del nstr[b]
    p = p + 1
  return nstr

df_tweet['Preprocessing'] = df_tweet['Tweet'].apply(token)
df_tweet
```

```python
#Proses Stopword Removal / Filtering
! pip install Sastrawi
nltk.download('stopwords')
from nltk.corpus import stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

def stopword_removal(preprocessing):
  filtering = StopWordRemoverFactory().get_stop_words()
  x = []
  data = []
  def myFunc(x):
    if x in filtering:
      return False
    else:
      return True
  fit = filter(myFunc, preprocessing)
  for x in fit:
    data.append(x)
  return data

df_tweet['Preprocessing'] = df_tweet['Preprocessing'].apply(stopword_removal)
df_tweet
```

```python
#Stemming
! pip install Sastrawi
from sklearn.pipeline import Pipeline
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(preprocessing):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for w in preprocessing:
    dt = stemmer.stem(w)
    do.append(dt)
  d_clean = []
  d_clean = " ".join(do)
  print(d_clean)
  return d_clean

df_tweet['Preprocessing'] = df_tweet['Preprocessing'].apply(stemming)
```
